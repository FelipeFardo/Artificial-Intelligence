```python
# Lista de Targets
targets = ['Risco_Credito', 'Risco_Compliance', 'Risco_Mercado']
```

```python
# Visualizar Distribuição dos targets
for col in targets:
  df_count = df_companies[col].value_counts().reset_index()
  df_count.columns = ['categoria', 'contagem']

  fig = px.bar(df_count,
               x="categoria",
               y='contagem',
               title=f'Distribuição/ Contagem de {col}'
               )
  
  fig.show()
```

```python
# Analisar relação entre featurs numéricas e categoricas e os targets
for target in targets:
  for col in numeric_features:
    fig = px.box(df_companies, x=target, y=col, title=f'{col} por {target}')
    fig.show()

  for col in categorical_features:
    fig = px.histogram(df_companies, x=col, color=target, barmode='group',title=f'{col} por {target}')
    fig.show()
```

```python
# Matriz de Correlação
correlation_matriz = df_companies[numeric_features.tolist()+targets].corr()

# Heatmap da matriz de correlação
fig = px.imshow(correlation_matriz,
                color_continuous_scale='armyrose',
                title='Matriz de Correlação',
                zmin=-1,
                zmax=1
                )

# Ajustes no Heatmap
fig.update_traces(text=correlation_matriz, texttemplate='%{text:.1%}', textfont=dict(size=9))

fig.update_layout(
  width=1000,
  height=600,
  title_font=dict(size=14),
  font=dict(size=10)
)

fig.show()
```

```python
# ANOVA - Análise de Variância (diferença significativa entre média entre 3 ou mais grupos)
# T - Student - (diferença significativa entre média entre 2 grupos)
for col in numeric_features:
  for target in targets:
    groups = [df_companies[df_companies[target]==val][col] for val in df_companies[target].unique()]
    if df_companies[target].nunique()<=2:
      stat, p = ttest_ind(*groups)
      test = 'T-Test'
    elif df_companies[target].nunique()>2:
      stat, p = f_oneway(*groups)
      test = 'ANOVA'

    print(f'{Fore.RED if p<0.05 else Fore.WHITE}'
            f'{test} entre {col} e {target}: p-valor = {p}')
```

### Preparação dos dados

```python
# Separar X e y
coluns_remove_indexs = ['ID', 'Nome_Empresa']

X = df_companies.drop(columns=coluns_remove_indexs + targets, axis=1)
y = df_companies[targets]
```

```python
# Separar entre Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=51, shuffle=True)
```

### Treinamento do Modelo Catboost com abordagem OnevsRest (OnevsAll)

```python
# Criar o objeto Catboost

cat_model = CatBoostClassifier(
  iterations=1000,
  learning_rate=0.05,
  depth=6,
  l2_leaf_reg=3,
  cat_features=categorical_features.tolist(),
  random_strength=2,
  loss_function='Logloss'
)
```

```python
# Colocar o modelo Catboost na abordagem OneVsRest
ovr_model = OneVsRestClassifier(cat_model)
```

```python
# Treinar o modelo
ovr_model.fit(X_train, y_train)
```

### Análise dos Resultados e métricas

```python
# Fazer predição das classes com base no modelo
y_pred = ovr_model.predict(X_test)
y_pred
```

```python
# Fazer a predição das probabilidades com base no modelo
y_pred_proba = ovr_model.predict_proba(X_test)
y_pred_proba
```

```python
# Relatório de classificação
print('Relatório de Classificação')
print(classification_report(y_test,y_pred))
```

```python
# Mostrar a Matriz de cunfusão para cada classe
for i, target in enumerate(targets):
  print(f'\nMatriz de Confusão para {target}')
  conf_matrix = confusion_matrix(y_test.iloc[:,1], y_pred[:,1])
  fig = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=[0,1])
  fig.plot()
  plt.title(F'Matriz de confusão para {target}')
  plt.show()
```

```python
# Calcular o logloss para cada target
print('Log Loss por Target')

for i, target in enumerate(targets):
  log_loss_score = log_loss(y_test.iloc[:,1], y_pred_proba[:,1])
  print(f'{target}: {log_loss_score:.4f}')
```

### Interpretabilidade

```python
# Importância das variáveis - Para cada target

for i, target in enumerate(targets):
  # Obter a importância das features
  importances = ovr_model.estimators_[i].feature_importances_
  features_names =  X.columns.tolist()

  # Cria um Dataframe com as importâncias e nomes de features
  df_importances = pd.DataFrame({'feature': features_names, 'importance': importances})

  # Ordenar o Dataframe
  df_importances = df_importances.sort_values('importance', ascending=False).reset_index(drop=True)

  # Montar o Plot
  fig = px.bar(
    df_importances,
    x='importance',
    y='feature',
    orientation='h',
    title=f'Importãncia das 10 principais features - {target}',
    labels={'importance': 'Importância', 'feature': 'Feature'},
    color='importance',
    color_continuous_scale='ArmyRose'
  )
  
  # Personalizar o layout

  fig.update_layout(
    height=600,
    yaxis={'categoryorder': 'total ascending'}
  )

  fig.show()
```

```python
# Visualizar exemplo de árvore

model_catboost = ovr_model.estimators_[0]

# Criar o Pool com os dados de treinamento
pool = Pool(
  data=X_train,
  label=y_train.iloc[:,0],
  cat_features=categorical_features.tolist(),
  feature_names=list(X_train.columns)
)

model_catboost.plot_tree(tree_idx=0,pool=pool)
```

```python
# Shap Values

# Criar e rodar explainers para acada target
explainers = []
shap_values = []

for i, target in enumerate(targets):
  print(f'Calculando SHAP Values para {target}...')

  # Usar TreeExplainer específico pra Catboost
  explainer = shap.TreeExplainer(ovr_model.estimators_[i])

  shap_values_target = explainer.shap_values(X_test)
  
  # Atualizo as listas
  explainers.append(explainer)
  shap_values.append(shap_values_target)
```

```python
# Plotar gráficos SHAP para cada target

for i, target in enumerate(targets):
  print(f'\nGráfico SHAP para {target}:')

  # Criar o gráfico resumo de SHAP
  shap.summary_plot(
    shap_values[i],
    X_test,
    plot_type='bar',
    title=f'Contribuição Global das Variáveis - {target}',
    show=False
  )

  plt.tight_layout()
  plt.show()

  # Criar o gráfico resumo de SHAP
  shap.summary_plot(
    shap_values[i],
    X_test,
    title=f'Distribuição dos Valores SHAP - {target}',
    show=False
  )

  plt.tight_layout()
  plt.show()
```
